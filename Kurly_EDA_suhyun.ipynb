### 그러면 할인율과 월간 판매량을 분석해보자
TODO: 월간 판매량이 아니라 월간 리뷰 갯수이다. 여기서부터 오인이 시작된듯

```python
discount_df = pd.DataFrame([df['상품명'],df['판매가'],df['할인가'],df['할인율'],df['쿠폰여부'],df['쿠폰할인율'], df['year_month']]).T
discount_df
```

```python
print(len(discount_df[discount_df['할인율']!= 0]))
print(len(discount_df['쿠폰할인율'] != 0))
```

```python
plt.hist(df['할인율'], bins = 20)
plt.title('할인율별 리뷰 갯수')
plt.xlabel('할인율')
plt.ylabel('상품의 구매에 따른 리뷰 갯수')
print('전체 리뷰 중 할인상품에 대한 리뷰 갯수는', len(df[df['할인율']!=0])*100/len(df),'% 이다')
```

```python
discount_rate_y = np.array([1660151.,       0.,       0.,       0.,       0.,   53929.,
              0.,   15723.,       0.,   29618.,   38082.,       0.,
              0.,  157471.,       0.,       0.,       0.,    8747.,
              0.,   19502.])

discount_rate_x = np.array([ 0. ,  1.4,  2.8,  4.2,  5.6,  7. ,  8.4,  9.8, 11.2, 12.6, 14. ,
        15.4, 16.8, 18.2, 19.6, 21. , 22.4, 23.8, 25.2, 26.6])

discount_rate_y_per = list(map(lambda x:x*100/len(df), discount_rate_y))
plt.bar(discount_rate_x, discount_rate_y_per)
plt.title('할인율별 리뷰 갯수 비율')
plt.xlabel('할인율')
plt.ylabel('상품의 구매에 따른 리뷰 갯수 비율')
```

```python
disc_month = {key : [0,0] for key in list(df['year_month'].unique())}
#key : 해당 연월, value[0] : 할인된 상품 리뷰 갯수, value[1] : 할인안된 상품 리뷰 갯수
```

```python
for row in discount_df.itertuples():
    if row.할인율 == 0: disc_month[row.year_month][1] += 1
    else: disc_month[row.year_month][0] += 1
```

```python
y_m = np.array(list(disc_month.keys()))
discounted = np.array([i[0] for i in disc_month.values()])
non_discounted = np.array([i[1] for i in disc_month.values()])
ratio = []
for i in range(len(y_m)):
    ratio.append(discounted[i]*100/(discounted[i]+non_discounted[i]))
```

```python
fig, ax = plt.subplots(2,1, figsize = (30,14))
idx = np.arange(len(y_m))
w = 0.35

ax[0].bar(idx-w/2, discounted, color = 'royalblue', width = w, label = 'discounted')
ax[0].bar(idx+w/2, non_discounted, color = 'orange', width = w, label = 'original price')
ax[0].set_xticks(idx)
ax[0].set_xticklabels(y_m)
ax[0].legend()
ax[0].set_title('월별 할인/비할인상품 리뷰 등록 갯수')
ax[0].set_xlabel('연-월')
ax[0].set_ylabel('리뷰 갯수')

ax[1].bar(y_m, ratio, color = 'royalblue')
ax[1].set_title('월별 할인상품 리뷰 갯수의 비율')
ax[1].set_xlabel('연-월')
ax[1].set_ylabel('할인상품 리뷰 비율')

plt.show()
```
