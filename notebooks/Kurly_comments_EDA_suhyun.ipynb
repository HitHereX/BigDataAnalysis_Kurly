### 시기에 따른 할인상품 갯수
22년도에는 할인상품 리뷰 갯수가 비슷하고 비율도 큰 변동이 없으나 21년도에는 할인상품 리뷰 갯수가 증가하는 시기가 있으며, 이때 리뷰갯수도 증가하였으므로 매출이 신장되었음을 예상할 수 있다

```python
#making feature
disc_month = {key : [0,0] for key in list(df['year_month'].unique())}

for row in discount_df.itertuples():
    if row.할인율 == 0: disc_month[row.year_month][1] += 1
    else: disc_month[row.year_month][0] += 1
```

```python
y_m = np.array(list(disc_month.keys()))
discounted = np.array([i[0] for i in disc_month.values()])
non_discounted = np.array([i[1] for i in disc_month.values()])
ratio = []
for i in range(len(y_m)):
    ratio.append(discounted[i]*100/(discounted[i]+non_discounted[i]))
```

```python
fig, ax = plt.subplots(2,1, figsize = (30,14))
idx = np.arange(len(y_m))
w = 0.35

ax[0].bar(idx-w/2, discounted, color = 'royalblue', width = w, label = 'discounted')
ax[0].bar(idx+w/2, non_discounted, color = 'orange', width = w, label = 'original price')
ax[0].set_xticks(idx)
ax[0].set_xticklabels(y_m)
ax[0].legend()
ax[0].set_title('월별 할인/비할인상품 리뷰 등록 갯수')
ax[0].set_xlabel('연-월')
ax[0].set_ylabel('리뷰 갯수')

ax[1].bar(y_m, ratio, color = 'royalblue')
ax[1].set_title('월별 할인상품 리뷰 갯수의 비율')
ax[1].set_xlabel('연-월')
ax[1].set_ylabel('할인상품 리뷰 비율')

plt.show()
```

### 단어 빈도 분석
sklearn package의 CountVectorizer 를 이용하여 리뷰내용의 n-gram에서 많이 등장하는 단어를 시각화 했다.

아래 product_comment_dict.keys() 의 결과를 참조하여 상품명을 PRODUCT 변수에 넣으면 원하는 상품의 빈출 상위 n개 단어의 빈도수와 비율을 알 수 있다.

```python
cv = CountVectorizer(ngram_range = (1,2))
words = cv.fit_transform(df['리뷰내용'])
sum_words = words.sum(axis = 0)

words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)
frequency = pd.DataFrame(words_freq, columns=['word', 'freq'])

frequency.head(20).plot(x='word', y='freq', kind='bar', figsize=(15, 6))
plt.title("Most Frequently Occuring Words - Top 20")
plt.show()
```

```python
product_comment_dict = {key : [] for key in list(df['상품명'].unique())}
for row in df.itertuples():
    product_comment_dict[row.상품명].append(row.리뷰내용)
```

```python
product_comment_dict.keys()
```

```python
PRODUCT = "[비비고] 노릇하게 구운 가자미 구이"
HEADS = 5

cv = CountVectorizer(ngram_range = (1,2))
words = cv.fit_transform(product_comment_dict[PRODUCT])
sum_words = words.sum(axis = 0)

words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)
frequency = pd.DataFrame(words_freq, columns=['word', 'freq'])

## BAR PLOT
frequency.head(HEADS).plot(x='word', y='freq', kind='bar')
plt.title("Most Frequently Occuring Words of %s - Top %d" %(PRODUCT, HEADS))
plt.show()


# PIE PLOT
ratio = []
labels = []
for i in range(HEADS):
    ratio.append(words_freq[i][1])
    labels.append(words_freq[i][0])

plt.title("Most Frequently Occuring Words of %s - Top %d" %(PRODUCT, HEADS))
plt.pie(ratio, labels = labels, autopct = '%.0f%%')
plt.show()
```
