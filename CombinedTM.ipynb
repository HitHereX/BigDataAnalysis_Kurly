```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
import re
```

```python
import pyLDAvis as vis
from contextualized_topic_models.utils.preprocessing import WhiteSpacePreprocessing
from contextualized_topic_models.models.ctm import CombinedTM
from contextualized_topic_models.utils.data_preparation import TopicModelDataPreparation, bert_embeddings_from_list
```

```python
#get corpus from raw data
df = pd.read_csv('hehe.csv')
prepped = []
for line in df["리뷰 내용"]:
    line=str(line)
    # 빈 문자열이거나 숫자로만 이루어진 줄은 제외
    if line and not line.replace(' ', '').isdecimal():
        prepped.append(line)
```

```python
class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
    def __call__(self, sent):
        word_tokens = self.tagger.morphs(sent)
        result = [word for word in word_tokens if len(word) > 1]
        return result
```

```python
custom_tokenizer = CustomTokenizer(Okt())
```

```python
vectorizer = CountVectorizer(tokenizer=custom_tokenizer, 
                             max_features=3000)
```

```python
train_bow_embeddings = vectorizer.fit_transform(prepped)
```

```python
vocab = vectorizer.get_feature_names_out()
id2token = {k: v for k, v in zip(range(0, len(vocab)), vocab)}
```

```python
#터지면 batch_size 줄이고, 리소스 보고 gpu ram 자리 있으면 더 키우고(2의 n승 단위)
train_contextualized_embeddings = bert_embeddings_from_list(prepped,#if burst decrease, 리소스 보고 gpu ram 자리 있으면 더 키우고 
                                                            "sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens",
                                                            batch_size = 2048, 
                                                            max_seq_length = 100)
```

```python
qt = TopicModelDataPreparation()
training_dataset = qt.load(train_contextualized_embeddings, train_bow_embeddings, id2token)
```

```python
ctm = CombinedTM(bow_size=len(vocab), 
                 contextual_size=768, 
                 n_components=8, 
                 num_epochs=2) #되는지 테스트만 하려면 2, 결과를 보려면 20+

ctm.fit(training_dataset)
```

```python
lda_vis_data = ctm.get_ldavis_data_format(vocab, 
                                          training_dataset, 
                                          n_samples=10)
```

```python
ctm_pd = vis.prepare(lda_vis_data['topic_term_dists'], 
                     lda_vis_data['doc_topic_dists'],
                     lda_vis_data['doc_lengths'],
                     lda_vis_data['vocab'],
                     lda_vis_data['term_frequency'],)
vis.display(ctm_pd)
```
